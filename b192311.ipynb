```python
import pandas as pd
```

```python
df = df.dropna()
```

```python
print(df.shape)
```

```python
rooms=df["Regionname"]
```

```python
price = df["Price"]
```

```python
import matplotlib.pyplot as plt
```

```python
print(min(price))
print(max(price))
print(max(price)-min(price))
```

```python
fig = plt.figure()
plt.hist(x=price,edgecolor = "green",bins=40)
plt.xlabel("Price range")
plt.ylabel("Number of houses")
plt.title("Histogram of price")
plt.show()
```

```python
rooms_count=df["Regionname"].value_counts()
```

```python
rooms_count = dict(rooms_count)
```

```python
print(rooms_count)
```

```python
a=list(rooms_count.keys())
b=list(rooms_count.values())
```

```python
fig = plt.figure(figsize=(8,8))

plt.pie(x=b,labels=a)
plt.legend()

plt.show()
```

```python
method_count = df["Method"].value_counts()
method_count = dict(method_count)
print(method_count)
```

```python
type_count = df["Type"].value_counts()
type_count = dict(type_count)
print(type_count)
```

```python
suburb_count = df["Suburb"].value_counts()
suburb_count = dict(suburb_count)
print(suburb_count)
```

```python
fig = plt.figure(figsize=(10,10))

plt.bar(x=k,height=v)


plt.show()
```

```python
price=df["Price"]
price=list(price)
distance=df["Distance"]
distance=list(distance)
```

```python
fig = plt.figure(figsize=(10,10))
plt.plot(distance,price,linestyle="",marker="*",color="green",markerfacecolor="red")
plt.xlabel("Distance")
plt.ylabel("Price")
plt.title("Distance Vs Price")
plt.show()
```
